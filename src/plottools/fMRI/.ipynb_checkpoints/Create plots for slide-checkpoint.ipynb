```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
sns.set()
```

```python
data = np.array(
    [[5, 2, 4, 1, 3, 2],
     [1, 4, 1, 4, 1, 4],
     [3, 1, 5, 2, 5, 2],
     [1, 4, 1, 5, 3, 4],
     [3, 1, 4, 2, 3, 2],
     [2, 4, 2, 4, 2, 4]])

users = ['James', 'Mary', 'Robert', 'Linda', 'David', 'Susan']
movies = ['Aloha', 'Star Wars', 'American Pie', 'Hunger games', 'Silver Linings', 'Maze Runner']

df = pd.DataFrame(data, index=users, columns=movies )
```

```python

%matplotlib inline
cmap = sns.diverging_palette(h_neg=5, h_pos=216, s=65, l=58, sep=34, n = 13)

sns.set_context("talk", font_scale=1.4)
plt.rcParams["font.family"] = "pt sans"

f, ax = plt.subplots(figsize=(6,6))
matix_ax = sns.heatmap(df, annot=True, cbar=False, linewidths=0, ax=ax, cmap=cmap, center=1)

matix_ax.xaxis.set_ticks_position('top')
for tick in matix_ax.get_xticklabels():
    tick.set_rotation(45)
    tick.set_ha('left')
for tick in matix_ax.get_yticklabels():
    tick.set_rotation(0)
```

```python
sns.choose_cubehelix_palette()
```

```python
u, s, vh = np.linalg.svd(data, full_matrices=True)
v = vh.T
```

```python

color_palette = sns.diverging_palette(220, 20, n=7)
min_val = np.min(np.concatenate([u, vh]))
max_val = np.max(np.concatenate([u, vh]))

for i in [0, 1]:
    f, ax1 = plt.subplots(figsize=(1, 6))
    u_i = pd.DataFrame(u[i], columns = [f'$u_{i+1}$'])

    u_ax = sns.heatmap(u_i, annot=True, cbar=False, yticklabels=False,  xticklabels=False, fmt='.1f', 
                       linewidths=0, cmap=cmap, center=0,  vmin=min_val, vmax=max_val)
    for tick in u_ax.get_yticklabels():
        tick.set_rotation(0)
    f.show()
        
```

```python

for i in [0, 1]:
    f, ax2 = plt.subplots(figsize=(6, 1))
    v_i = pd.DataFrame(vh[:,i])

    v_ax = sns.heatmap(v_i.T, annot=True, cbar=False, xticklabels=False, yticklabels=False, fmt='.1f', 
                       linewidths=0, cmap=cmap, center=0,  vmin=min_val, vmax=max_val)
    
    f.show()
```
