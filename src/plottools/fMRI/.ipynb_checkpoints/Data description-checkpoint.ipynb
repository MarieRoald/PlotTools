```python
from pathlib import Path

import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

import ipywidgets as widgets

import h5py

%matplotlib notebook
```

```python
%matplotlib notebook
fig, sub = plt.subplots()

@widgets.interact(source=['IA3T', 'IOWA', 'MAGH', 'MINN', 'NMEX'])
def choose_source(source):
    source = Path('/media/marie/My Passport/DATASETS/MCIC')/source
    patients = sorted(source.iterdir())
    @widgets.interact(patient=patients)
    def choose_patient(patient):
        images_dir = patient/'month000'/'sm'
        
        @widgets.interact(imagefile=images_dir.iterdir())
        def choose_image(imagefile):
            imagefile = imagefile.iterdir().__next__()
            img = nib.load(str(imagefile))
            print(img.header)
            data = img.get_fdata()
            data -= data.min()
            data /= data.max()
            
            @widgets.interact(i=(0, data.shape[2]-1), t=(0, data.shape[3]-1), gamma=(0.5, 2))
            def replot(i, t, gamma):
                sub.clear()
                sub.imshow(data[:, :, i, t]**gamma, vmin=0, vmax=1, cmap='inferno')
```
